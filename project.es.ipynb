#### Ejercicio 18. Normaliza la variable de precios para cada población y traza los 4 histogramas en el mismo gráfico (debes usar el subconjunto obtenido en la pregunta 14) (★★★)

Para el método de normalización, puedes usar el que consideres adecuado, no hay una única respuesta correcta para esta pregunta. Imprime el gráfico y escribe en la celda de Markdown un breve análisis sobre el gráfico.

Pista: Puedes ayudarte revisando la demostración multihist de Matplotlib.

```python
ds_cinturon_sur_1 = ds[ds['level5'].isin(poblaciones_cinturon_sur)].copy()

def normalizacion_max_min(series):
    return (series - series.min()) / (series.max() - series.min())

ds_cinturon_sur_1['price_normalized'] = ds_cinturon_sur_1.groupby('level5')['price'].transform(normalizacion_max_min)

plt.figure(figsize=(10, 6))
for poblacion in poblaciones_cinturon_sur:
    subset = ds_cinturon_sur_1[ds_cinturon_sur_1['level5'] == poblacion]
    plt.hist(subset['price_normalized'], bins=10, alpha=0.5, label=poblacion)

plt.title('Distribución Normalizada de Precios en el Cinturón Sur de Madrid', fontsize=16)
plt.xlabel('Precio Normalizado', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.legend(title='Población')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##### Análisis del Gráfico de Histogramas Normalizados

El gráfico muestra la distribución normalizada de los precios de las viviendas en las poblaciones del "cinturón sur" de Madrid: **Fuenlabrada**, **Leganés**, **Getafe** y **Alcorcón**.

1. **Getafe** tiene una distribución de precios más amplia, lo que indica una mayor variabilidad en los precios de las viviendas.
2. **Leganés**, **Alcorcón** y **Fuenlabrada** muestran distribuciones más concentradas (salvo por ciertos picos puntuales), lo que sugiere que los precios en estas poblaciones son más homogéneos.

#### Ejercicio 20. Realiza el mismo gráfico para 4 poblaciones diferentes (columna level5) y colócalos en el mismo gráfico. Debes usar el subconjunto obtenido en la pregunta 14 (★★☆) 
Pista: Haz un diagrama de dispersión de cada población usando subgráficos (subplots).

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Relación entre Precio y Superficie en el Cinturón Sur de Madrid', fontsize=14)

colores = {'Fuenlabrada': 'blue', 'Leganés': 'green', 'Getafe': 'red', 'Alcorcón': 'purple'}

for i, poblacion in enumerate(poblaciones_cinturon_sur):
    ax = axes[i // 2, i % 2]
    
    subset = ds_cinturon_sur[ds_cinturon_sur['level5'] == poblacion]
    
    ax.scatter(subset['surface'], subset['price'], color=colores[poblacion], label=poblacion, alpha=0.7)
    
    ax.set_title(poblacion, fontsize=10)
    ax.set_xlabel('Superficie (m²)', fontsize=12)
    ax.set_ylabel('Precio (€)', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.7)
    ax.legend()

plt.tight_layout()
plt.show()
```

Ejecuta la siguiente celda y luego comienza a codear en la siguiente. Debes implementar un código simple que transforme las columnas de coordenadas en un diccionario de Python (agrega más información si es necesario) y agrégala al mapa.

```python
from ipyleaflet import Map, basemaps, Marker

# Mapa centrado en (60 grados latitud y -2.2 grados longitud)
# Latitud, longitud
map = Map(center=(60, -2.2), zoom=2, min_zoom=1, max_zoom=20, basemap=basemaps.OpenStreetMap.Mapnik)
```
